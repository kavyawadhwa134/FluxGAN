```python
import csv
import matplotlib.pyplot as plt

# Correct path to your CSV file
file_path = '../plots/loss_log.csv'

iterations = []
discriminator_losses = []
generator_losses = []

try:
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            iterations.append(int(row['Iteration']))
            discriminator_losses.append(float(row['Discriminator Loss']))
            generator_losses.append(float(row['Generator Loss']))

    # Plot the Discriminator and Generator Loss
    plt.plot(iterations, discriminator_losses, label='Discriminator Loss')
    plt.plot(iterations, generator_losses, label='Generator Loss')

    plt.legend()
    plt.title('Training Losses')
    plt.tight_layout()
    plt.savefig('../plots/training_loss.png')  # Save to the ../plots directory
    plt.show()

except FileNotFoundError:
    print(f"File not found: {file_path}")
except KeyError as e:
    print(f"Missing expected column in the CSV file: {e}")
```
